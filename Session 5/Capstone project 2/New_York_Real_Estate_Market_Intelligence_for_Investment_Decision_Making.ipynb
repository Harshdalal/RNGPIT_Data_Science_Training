#Project Title

New York Real Estate Market Intelligence for Investment Decision Making

# Source Link 🇰

https://www.kaggle.com/datasets/kanchana1990/new-york-real-estate-data-2026


#Dataset:
New York Real Estate Data 2026 – contains property characteristics such as location, property type, bedrooms, bathrooms, price, broker information, and geographic attributes useful for housing market analysis.

# Scenario

You are a Data Analyst working for a real estate investment firm.

The firm wants to understand:

> Which property segments offer the best investment potential

> Which locations have the highest price growth

> Which property characteristics influence pricing

> Whether pricing differences across boroughs and property types are statistically significant

Your job is to analyze the dataset and produce insights for investors.

#Phase 1 — Data Engineering (20 min)
Objective

Prepare the dataset for analysis.

Students must:

> Load dataset

> clean data

> inspect schema

> transform columns

#Business Tasks

1. The firm wants to build a market intelligence dashboard.

Before doing that:

> Identify key variables describing a property listing

> Identify missing values or inconsistent entries

2. Some properties have missing bedroom or bathroom information.

How should the firm handle these records to maintain reliable analysis?

3. Convert any columns that represent:

> price

> dates

> geographic values

into correct formats.

Skills Tested

> Data engineering

> Pandas

> NumPy

#Phase 2 — Exploratory Data Analytics (30 min)
#Business Objective

Understand overall structure of the New York housing market.

#Questions

1. Which boroughs or neighborhoods contain the highest concentration of property listings?

2. What price ranges dominate the New York real estate market?

3. Which property types appear most frequently in listings?

4. Do luxury properties concentrate in specific areas of New York?

5. What property features appear most common among high-value listings?

#Required Charts

Students must create at least:

> Histogram → property price distribution

> Bar chart → property types

> Count plot → borough listings

> Box plot → price by location

Libraries:

> Seaborn

> Matplotlib

#Phase 3 — Visualization & Market Intelligence (20 min)

#Objective

Real estate executives want interactive visual insights.

Students must use Plotly.

#Tasks

1. Create interactive price distribution by borough.

2. Build a geographic visualization of listings.

3. Create a price vs property size interactive chart.

4. Build an interactive chart comparing property types across boroughs.

#Skills Tested

> Plotly

> advanced visualization

> storytelling with data

#Phase 4 — Probability & Market Risk Analysis (20 min)
#Business Question

Investors want to estimate probability of high-value investment opportunities.

Students must analyze:

1. What is the probability that a randomly selected property exceeds a premium price threshold?

2. How does the likelihood of high-value properties vary across boroughs?

3. If a property has more bedrooms and larger space, how does the probability of high price change?

Students should compute probabilities and interpret them in investment terms.

#Phase 5 — Statistical Market Validation (20 min)

Students must determine whether observed differences are statistically meaningful.

#Business Question 1

Investors suspect:

> Properties in Manhattan are significantly more expensive than properties in other boroughs.

Students must verify whether the difference in average price is statistically significant.

(Expected method: t-test)

#Business Question 2

Investment analysts believe:

> Property prices vary significantly across different boroughs of New York.

Students must evaluate whether location significantly impacts property pricing.

(Expected method: ANOVA test)

#Business Question 3

Real estate regulators want to understand:

Whether property type and listing price category are related.

Students must analyze whether property type influences price category.

(Expected method: Chi-Square test)

#Phase 6 — Business Insight & Decision Making (20 min)

Students must prepare strategic recommendations.

Questions

1. Which borough appears most attractive for real estate investment?

2. Which property features drive higher market prices?

3. Are luxury properties concentrated in certain neighborhoods?

4. Which market segments appear undervalued or emerging?

5. What data-driven strategy should the investment firm adopt?
