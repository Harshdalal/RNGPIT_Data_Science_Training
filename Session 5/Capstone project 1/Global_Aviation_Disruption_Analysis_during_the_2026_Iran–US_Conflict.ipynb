#Project Title

Global Aviation Disruption Analysis during the 2026 Iran–US Conflict

#Scenario

You are working as a Data Analyst for an international aviation monitoring organization.
The organization wants to understand how geopolitical conflicts disrupt global aviation operations.

Your task is to analyze the dataset and produce insights for airline companies, airport authorities, and policy makers.

# Source Link :
https://www.kaggle.com/datasets/zkskhurram/global-civil-aviation-disruption2026-iranus-war/data

#Phase 1 — Data Engineering & Data Understanding (20 min)

Students must first prepare the dataset.

**Tasks**

1.Load the dataset into Pandas.

2.Identify dataset structure.

#Questions:

> What are the main entities represented in the dataset?

> Which columns represent time, location, airline operations, or disruptions?

> Are there missing or inconsistent values?

> If some columns contain text dates or numeric strings, how would you convert them?

#Technical Requirements

Students must use:

> pandas

> numpy

Expected operations:

> missing value detection

> type conversion

> basic dataset profiling

#Phase 2 — Exploratory Data Analytics (30 min)

Business Objective

Airlines want to understand how severe the disruption was and where it occurred most frequently.

#Questions

> Which countries or regions experienced the highest aviation disruptions?

> Which airlines or airports were most affected?

> What time period shows the highest disruption activity?

> Are disruptions concentrated in specific flight corridors or airspace zones?

#Required Visualizations

> Heatmap

#Libraries:

#Phase 4 — Probability & Risk Analysis (20 min)

Business Question

Airlines want to estimate risk levels during conflict periods.

#Students must analyze:

1.What is the probability that a flight operating in a conflict region will be disrupted?

2.How does disruption probability differ between regions or airlines?

3.If a flight passes through multiple high-risk regions, how does the probability of disruption change?

#Students should:

> calculate probabilities

>interpret operational risk

#Phase 5 — Statistical Testing (20 min)

Students must choose appropriate statistical tests to answer the following business questions.

#Question 1 — Operational Cost Analysis

Airlines suspect that flight delays increased significantly during the conflict period.

#Students must test:

Is there a statistically significant difference in flight delays before vs during disruption periods?

(Expected statistical approach: t-test)

#Question 2 — Regional Disruption Analysis

Airline regulators want to know:

Do different regions experience significantly different levels of aviation disruption?

(Expected statistical approach: ANOVA)

#Question 3 — Airline Risk Analysis

Aviation authorities want to understand:

Is there a relationship between airline identity and disruption occurrence?

(Expected statistical approach: Chi-square test)

#Phase 6 — Data Science Insight & Decision Making (20 min)

Students must produce executive-level insights.

#Business Questions

> Which global air routes are most vulnerable during geopolitical conflicts?

> How should airlines re-route flights to minimize disruption risk?

> Which airports should be considered backup hubs during crises?

> What early signals in the dataset indicate potential aviation crises?
